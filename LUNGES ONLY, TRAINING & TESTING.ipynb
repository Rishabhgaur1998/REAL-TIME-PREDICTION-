```python
sequences, labels = [], []
for action,videos in a_dict.items():
    for vid in videos:
        cap = cv2.VideoCapture(vid)
        filename = re.search('Camera Roll/Combines/(.+?).mp4',vid).group(1)  
        length = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
        print(vid)
        window = []
        for frame_num in range(lengthop):
            print(frame_num)
            print(vid)
            res = np.load(os.path.join(DATA_PATH, action, filename, "{}.npy".format(frame_num)))
            window.append(res)
        sequences.append(window)
        labels.append(label_map[action])
```

```python
np.array(sequences).shape
xx=np.array(sequences)
yy = to_categorical(labels).astype(int)
X_train, X_test, y_train, y_test = train_test_split(xx, yy, test_size=0.4,random_state=12)
```

```python
model3 = Sequential()
model3.add(LSTM(1024, return_sequences=True, activation='tanh', input_shape=(150,132)))
model3.add(LSTM(512, return_sequences=False, activation='tanh'))
model3.add(Dense(512, activation='tanh'))
model3.add(Dense(512, activation='tanh'))
model3.add(Dense(512, activation='tanh'))
model3.add(Dense(1024, activation='tanh'))
model3.add(Dense(len(a_dict.keys()), activation='softmax'))
optimizer_adam=Adam(learning_rate=0.000001,beta_1=0.9,beta_2=0.99) 
model3.compile(optimizer="sgd",loss='categorical_crossentropy',metrics=['categorical_accuracy'])

EarlyStop=EarlyStopping(patience=15,restore_best_weights=True)
LR=ReduceLROnPlateau(monitor='val_accuracy',verbose=1,factor=0.5,min_lr=1e-6)
check=ModelCheckpoint('minimal_LSTM_layer_model.hdf5',monitor='val_loss',verbose=1,save_best_only=True)
callback=[EarlyStop, LR, check]



import time

# starting time
start = time.time()


# history = model1.fit(X_train, y_train, batch_size=8, epochs=50,
#                     validation_split=0.1, verbose=2, callbacks=[check])

model3.fit(X_train, y_train, batch_size=8, epochs=10,
                    validation_split=0.1, verbose=2, callbacks=[check])
# end time
end = time.time()

# total time taken
print(f"Runtime of the model training is  is {(end - start)//60}")
```
